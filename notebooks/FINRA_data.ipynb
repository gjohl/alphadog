```python
import pandas as pd
import cufflinks as cf; cf.go_offline()

from alphadog.data.finra_data import load_finra_data
```

```python
start_1 = '20191104'
end_1 = '20191108'

start_2 = '20200325'
end_2 = '20200331'
```

```python
def get_finra_date_range(start, end):
    df_list = []
    
    for dt in pd.bdate_range(start, end):
        date_str = dt.strftime('%Y%m%d')
        tmp_df = load_finra_data(date_str)
        df_list.append(tmp_df)
        
    df = pd.concat(df_list)
    df = df.sort_values(['Date', 'Symbol'])
    
    return df
```

```python
old_df = get_finra_date_range(start_1, end_1)
new_df = get_finra_date_range(start_2, end_2)
```
